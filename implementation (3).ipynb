```python
#Importing libraries
import numpy as numpy
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
ds=pd.read_csv('Position_Salaries.csv')
ds
```

```python
#splitting dataset into Independent and dependent data

x=ds[["Level"]]
y=ds[["Salary"]]
```

```python
y
x
```

```python
# Plot graphs

plt.scatter(x,y)
plt.plot(x,y)
plt.show()
```

```python
# Model fitting(Polynomial regression model fitting)

from sklearn.linear_model import LinearRegression
from sklearn.preprocessing import PolynomialFeatures
poly_reg=PolynomialFeatures(degree=5)
x_poly=poly_reg.fit_transform(x)
x_poly
reg=LinearRegression()
reg.fit(x_poly,y)
```

```python
# visualization

plt.scatter(x,y)
plt.plot(x, reg.predict(poly_reg.transform(x)),color='red')
plt.show()
```

```python
# predicting for my input

myinput=pd.DataFrame(data=[[11]], columns=["Level"])
result=reg.predict(poly_reg.transform(myinput))
result
```
